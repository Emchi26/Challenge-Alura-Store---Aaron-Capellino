### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
for df in [tienda, tienda2, tienda3, tienda4]:
    df["Precio"] = pd.to_numeric(df["Precio"], errors="coerce")

print("üí∞ Facturaci√≥n total por tienda:")
print("Tienda 1:", tienda["Precio"].sum())
print("Tienda 2:", tienda2["Precio"].sum())
print("Tienda 3:", tienda3["Precio"].sum())
print("Tienda 4:", tienda4["Precio"].sum())
```

# 2. Ventas por categor√≠a

```python
print("üìä Categor√≠as m√°s populares por tienda:")

print("\nTienda 1:")
print(tienda["Categor√≠a del Producto"].value_counts())

print("\nTienda 2:")
print(tienda2["Categor√≠a del Producto"].value_counts())

print("\nTienda 3:")
print(tienda3["Categor√≠a del Producto"].value_counts())

print("\nTienda 4:")
print(tienda4["Categor√≠a del Producto"].value_counts())
```

# 3. Calificaci√≥n promedio de la tienda


```python
for df in [tienda, tienda2, tienda3, tienda4]:
    df["Calificaci√≥n"] = pd.to_numeric(df["Calificaci√≥n"], errors="coerce")

print("‚≠ê Calificaci√≥n promedio por tienda:")
print("Tienda 1:", tienda["Calificaci√≥n"].mean())
print("Tienda 2:", tienda2["Calificaci√≥n"].mean())
print("Tienda 3:", tienda3["Calificaci√≥n"].mean())
print("Tienda 4:", tienda4["Calificaci√≥n"].mean())
```

# 4. Productos m√°s y menos vendidos

```python
print("üèÜ Productos m√°s y menos vendidos por tienda:")

print("\nTienda 1 ‚Äì Top 5 m√°s vendidos:")
print(tienda["Producto"].value_counts().head(5))
print("\nTienda 1 ‚Äì Bottom 5 menos vendidos:")
print(tienda["Producto"].value_counts().tail(5))

print("\nTienda 2 ‚Äì Top 5 m√°s vendidos:")
print(tienda2["Producto"].value_counts().head(5))
print("\nTienda 2 ‚Äì Bottom 5 menos vendidos:")
print(tienda2["Producto"].value_counts().tail(5))

print("\nTienda 3 ‚Äì Top 5 m√°s vendidos:")
print(tienda3["Producto"].value_counts().head(5))
print("\nTienda 3 ‚Äì Bottom 5 menos vendidos:")
print(tienda3["Producto"].value_counts().tail(5))

print("\nTienda 4 ‚Äì Top 5 m√°s vendidos:")
print(tienda4["Producto"].value_counts().head(5))
print("\nTienda 4 ‚Äì Bottom 5 menos vendidos:")
print(tienda4["Producto"].value_counts().tail(5))
```

# 5. Env√≠o promedio por tienda

```python
for df in [tienda, tienda2, tienda3, tienda4]:
    df["Costo de env√≠o"] = pd.to_numeric(df["Costo de env√≠o"], errors="coerce")

print("üöö Costo de env√≠o promedio por tienda:")
print("Tienda 1:", tienda["Costo de env√≠o"].mean())
print("Tienda 2:", tienda2["Costo de env√≠o"].mean())
print("Tienda 3:", tienda3["Costo de env√≠o"].mean())
print("Tienda 4:", tienda4["Costo de env√≠o"].mean())
```

# 6. Gr√°ficos

Gr√°fico de barras: Facturaci√≥n total por tienda

```python
import matplotlib.pyplot as plt

facturacion = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]
labels = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
colores = ["#4CAF50", "#2196F3", "#FFC107", "#E91E63"]

plt.figure(figsize=(8,5))
barras = plt.bar(labels, facturacion, color=colores, edgecolor="black")

# Etiquetas de valor encima de cada barra
for barra, valor in zip(barras, facturacion):
    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height() + (barra.get_height()*0.01),
             f"${valor:,.0f}", ha='center', fontsize=10, fontweight="bold")

# Formato y t√≠tulos
plt.title("Facturaci√≥n total por tienda", fontsize=14, fontweight="bold")
plt.ylabel("Facturaci√≥n ($)", fontsize=12)
plt.gca().set_yticklabels([f"${int(x):,}" for x in plt.gca().get_yticks()])

plt.show()
```

Gr√°fico de l√≠neas m√∫ltiples: Tendencia por tienda

```python
all_tiendas = pd.concat([tienda.assign(Tienda='Tienda 1'),
                         tienda2.assign(Tienda='Tienda 2'),
                         tienda3.assign(Tienda='Tienda 3'),
                         tienda4.assign(Tienda='Tienda 4')])

all_tiendas['Fecha de Compra'] = pd.to_datetime(all_tiendas['Fecha de Compra'], format='%d/%m/%Y')
all_tiendas['Mes'] = all_tiendas['Fecha de Compra'].dt.to_period('M')

ventas_por_mes_tienda = all_tiendas.groupby(['Mes', 'Tienda'])['Precio'].sum().reset_index()

ventas_por_mes_tienda['Mes'] = ventas_por_mes_tienda['Mes'].astype(str)

ventas_2022 = ventas_por_mes_tienda[ventas_por_mes_tienda['Mes'].str.startswith('2022')]

plt.figure(figsize=(12, 6))

for tienda_nombre, datos_tienda in ventas_2022.groupby('Tienda'):
    plt.plot(datos_tienda['Mes'], datos_tienda['Precio'], marker='o', label=tienda_nombre)

plt.title('Tendencia de Ventas por Tienda (2022)')
plt.ylabel('Ventas ($)')
plt.xlabel('Mes')
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x para mayor legibilidad
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Gr√°fico de barras - Calificaci√≥n promedio por tienda

```python
calificacion_promedio = [
    tienda["Calificaci√≥n"].mean(),
    tienda2["Calificaci√≥n"].mean(),
    tienda3["Calificaci√≥n"].mean(),
    tienda4["Calificaci√≥n"].mean()
]

plt.figure(figsize=(8,5))
barras = plt.bar(labels, calificacion_promedio, color=["#FF9800", "#4CAF50", "#2196F3", "#9C27B0"], edgecolor="black")

# Mostrar valores sobre las barras
for barra, valor in zip(barras, calificacion_promedio):
    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height() + 0.02,
             f"{valor:.2f}", ha='center', fontsize=10, fontweight="bold")

plt.title("Calificaci√≥n promedio por tienda", fontsize=14, fontweight="bold")
plt.ylabel("Calificaci√≥n promedio", fontsize=12)
plt.ylim(0, 5)  # Escala de 0 a 5
plt.show()
```
